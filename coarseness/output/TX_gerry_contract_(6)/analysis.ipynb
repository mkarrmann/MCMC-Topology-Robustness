```python
import json

with open("results.json") as f:
    results = json.load(f)
```

```python
unnested_results = [results[i*18:(i+1)*18] for i in range(0, len(results)//18)]
```

```python
for res in unnested_results:
    print(res)
```

```python
from statistics import mean
start = mean(res[0] for res in unnested_results)
print(start)
```

```python
diffs = []
for res in unnested_results:
    diff = list(map(lambda x: x - start, res))
    diffs.append(diff)
    print(diff)
```

```python
for i, diff in enumerate(diffs):
    print(i, diff[15])
```

```python
for i in range(len(diffs[0])):
    print(i * 500, max(diff[i] for diff in diffs))
```
